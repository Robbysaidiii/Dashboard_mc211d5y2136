## Exploratory Data Analysis (EDA)

### Explore `day`

```python
day.sample(5)
```

```python
day.describe(include="object")
```

```python
day.instant.is_unique
day.dteday.is_unique
```

```python
day.groupby(by="mnth").agg({
    "dteday": "nunique",
    "temp": ["max", "min", "mean", "std"]
})
```

```python
# Distribusi jumlah sewa sepeda (cnt)
sn.histplot(day['cnt'], bins=30, kde=True)
plt.title('Data penyewaan sepeda yang diaggregasi per hari (cnt)')
plt.xlabel('Jumlah Sewa Sepeda')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Korelasi antara suhu (temp) dan jumlah sewa sepeda (cnt)
sn.scatterplot(x=day['temp'], y=day['cnt'])
plt.title('Korelasi antara Suhu dan Jumlah Sewa Sepeda')
plt.xlabel('Suhu (Normalisasi)')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Visualisasi hubungan antara suhu dan jumlah sewa sepeda
sn.regplot(x=day['temp'], y=day['cnt'], scatter_kws={'alpha':0.3})
plt.title('Pengaruh Suhu terhadap Jumlah Sewa Sepeda')
plt.xlabel('Suhu (Normalisasi)')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

```python
# Rata-rata jumlah sewa sepeda pada hari kerja vs hari libur
avg_rentals = day.groupby('workingday')['cnt'].mean()
sn.barplot(x=avg_rentals.index, y=avg_rentals.values)
plt.title('Rata-rata Jumlah Sewa Sepeda: Hari Kerja vs Hari Libur')
plt.xlabel('Hari Kerja (1) / Hari Libur (0)')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda')
plt.show()
```

### Pertanyaan 2:

```python
# Visualisasi hubungan antara suhu dan jumlah sewa sepeda
sn.regplot(x=HOUR['temp'], y=HOUR['cnt'], scatter_kws={'alpha':0.3})
plt.title('Pengaruh Suhu terhadap Jumlah Sewa Sepeda')
plt.xlabel('Suhu (Normalisasi)')
plt.ylabel('Jumlah Sewa Sepeda')
plt.show()
```

```python
# Rata-rata jumlah sewa sepeda pada hari kerja vs hari libur
avg_rentals = HOUR.groupby('workingday')['cnt'].mean()
sn.barplot(x=avg_rentals.index, y=avg_rentals.values)
plt.title('Rata-rata Jumlah Sewa Sepeda: Hari Kerja vs Hari Libur')
plt.xlabel('Hari Kerja (1) / Hari Libur (0)')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda')
plt.show()
```
